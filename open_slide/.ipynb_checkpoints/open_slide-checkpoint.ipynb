### DeepZoomGenerator

```python
from openslide.deepzoom import DeepZoomGenerator
```

```python
img_gen = DeepZoomGenerator(img, tile_size=512, overlap=0, limit_bounds=False)
```

```python
## number of Deep Zoom levels in the image
img_gen.level_count
```

```python
## A list of (pixels_x, pixels_y) tuples for each Deep Zoom level. 
## level_dimensions[k] are the dimensions of level k
img_gen.level_dimensions
```

```python
## total number of Deep Zoom tiles in the image
img_gen.tile_count
```

```python
## a list of (tiles_x, tiles_y) tuples for each DeepZoomGenerator levels.
## level_tiles[k] are the tile counts of level k
img_gen.level_tiles
```

```python
## Return a string containing the XML metadata for the Deep Zoom .dzi file
## Parameters:format (str)  the delivery format of the individual tiles (png or jpeg)
img_gen.get_dzi('png')
```

```python
#Return an RGB Image for a tile.
#level (int):the Deep Zoom level
#address (tuple):  the address of the tile within the level as a (column, row) tuple

tile1 = img_gen.get_tile(11,(0,0))
tile2 = img_gen.get_tile(11,(0,1))

fig, ax = plt.subplots(1,2,figsize=(10,8))
ax[0].imshow(tile1)
ax[1].imshow(tile2)
```

```python
tile1.size
```
