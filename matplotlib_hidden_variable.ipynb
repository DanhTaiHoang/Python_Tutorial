```python
import numpy as np
import matplotlib.pyplot as plt

n0 = 100 ; g = 4. ; ln = 4

plt.rcParams.update({'font.size': 10})
plt.rcParams['font.family'] = ['Tahoma']
#['Tahoma', 'DejaVu Sans','Lucida Grande', 'Verdana']

nrow = 3
ncol = 4
scale_row = 4
scale_col = 1.8

plt.figure(figsize=(nrow*scale_row,ncol*scale_col))

# 1) =============================================================================================================
label1 = ['A','E']
label2 = ['B','F']
label3 = ['C','G']
label4 = ['D','H']

labelx = -2.5
labely = 1.8
def plot_w(nh0,ifig):
    print('nh0:',nh0)
    step1,step2,step3 = 5,1,1
    #step1,step2,step3 = 10,10,10
    #if nh0 == 10: step1,step2,step3 = 8,2,1
    #if nh0 == 40: step1,step2,step3 = 5,4,2    

    n = n0 - nh0
    ext_name = '%1.1f_%02d_%02d'%(g,nh0,ln)
    w0 = np.loadtxt('w0_%s.dat'%ext_name)

    wnMF = np.loadtxt('wfinal_%s_nMF.dat'%ext_name)
    wTAP = np.loadtxt('wfinal_%s_TAP.dat'%ext_name)
    weMF = np.loadtxt('wfinal_%s_eMF.dat'%ext_name)
    wMLE = np.loadtxt('wfinal_%s_MLE.dat'%ext_name)
    wFEM = np.loadtxt('wfinal_%s_FEM.dat'%ext_name)

    wmin = -1.8
    wmax = 1.8
    # obs to obs
    plt.subplot2grid((nrow,ncol),(ifig,0))
    if ifig ==0: plt.title('observed to observed')
    plt.text(labelx,labely,label1[ifig],fontsize=13)
    print(label1[ifig])
    plt.plot([wmin,wmax],[wmin,wmax],'--',linewidth=0.5,zorder=0)
    plt.scatter(w0[:n,:n:step1],wnMF[:n,:n:step1],marker='x',color='g',s=6,label='nMF',zorder=0)
    plt.scatter(w0[:n,:n:step1],wTAP[:n,:n:step1],marker='D',color='c',facecolors='none',s=6,label='TAP',zorder=1)
    plt.scatter(w0[:n,:n:step1],weMF[:n,:n:step1],marker='^',color='r',facecolors='none',s=6,label='eMF',zorder=2)
    plt.scatter(w0[:n,:n:step1],wMLE[:n,:n:step1],marker='o',color='b',facecolors='none',s=6,label='MLE',zorder=3)
    plt.scatter(w0[:n,:n:step1],wFEM[:n,:n:step1],marker='o',color='k',s=6,label='FEM',zorder=4)
    plt.xlabel('actual interactions')
    plt.ylabel('predicted interactions',labelpad=0)
    #ylab.set_position((0,0))
    plt.xlim([wmin,wmax])
    plt.ylim([wmin,wmax])        
    if ifig ==0: plt.legend(prop={'size': 7},labelspacing=0.25,loc='lower right',bbox_to_anchor=(1.,0.05))

    # hidden to obs
    plt.subplot2grid((nrow,ncol),(ifig,1))
    if ifig ==0: plt.title('hidden to observed')
    plt.text(labelx,labely,label2[ifig],fontsize=13)
    plt.plot([wmin,wmax],[wmin,wmax],'--',linewidth=0.5,zorder=0)
    plt.scatter(w0[:n:step2,n:],wnMF[:n:step2,n:],marker='x',color='g',s=6,label='nMF',zorder=0)
    plt.scatter(w0[:n:step2,n:],wTAP[:n:step2,n:],marker='D',color='c',facecolors='none',s=6,label='TAP',zorder=1)
    plt.scatter(w0[:n:step2,n:],weMF[:n:step2,n:],marker='^',color='r',facecolors='none',s=6,label='eMF',zorder=2)
    plt.scatter(w0[:n:step2,n:],wMLE[:n:step2,n:],marker='o',color='b',facecolors='none',s=6,label='MLE',zorder=3)
    plt.scatter(w0[:n:step2,n:],wFEM[:n:step2,n:],marker='o',color='k',s=6,label='FEM',zorder=4)
    plt.xlabel('actual interactions')
    plt.ylabel('predicted interactions',labelpad=0)
    plt.xlim([wmin,wmax])
    plt.ylim([wmin,wmax])

    # obs to hidden
    plt.subplot2grid((nrow,ncol),(ifig,2))
    if ifig ==0: plt.title('observed to hidden')
    plt.text(labelx,labely,label3[ifig],fontsize=13)
    plt.plot([wmin,wmax],[wmin,wmax],'--',linewidth=0.5,zorder=0)
    plt.scatter(w0[n:,:n:step2],wnMF[n:,:n:step2],marker='x',color='g',s=6,label='nMF',zorder=0)
    plt.scatter(w0[n:,:n:step2],wTAP[n:,:n:step2],marker='D',color='c',facecolors='none',s=6,label='TAP',zorder=1)
    plt.scatter(w0[n:,:n:step2],weMF[n:,:n:step2],marker='^',color='r',facecolors='none',s=6,label='eMF',zorder=2)
    plt.scatter(w0[n:,:n:step2],wMLE[n:,:n:step2],marker='o',color='b',facecolors='none',s=6,label='MLE',zorder=3)
    plt.scatter(w0[n:,:n:step2],wFEM[n:,:n:step2],marker='o',color='k',s=6,label='FEM',zorder=4)
    plt.xlabel('actual interactions')
    plt.ylabel('predicted interactions',labelpad=0)
    plt.xlim([wmin,wmax])
    plt.ylim([wmin,wmax])

    # hidden to hidden
    plt.subplot2grid((nrow,ncol),(ifig,3))
    plt.text(labelx,labely,label4[ifig],fontsize=13)
    if ifig ==0: plt.title('hidden to hidden')
    plt.plot([wmin,wmax],[wmin,wmax],'--',linewidth=0.5,zorder=0)
    plt.scatter(w0[n:,n::step3],wnMF[n:,n::step3],marker='x',color='g',s=6,label='nMF',zorder=0)
    plt.scatter(w0[n:,n::step3],wTAP[n:,n::step3],marker='D',color='c',facecolors='none',s=6,label='TAP',zorder=1)
    plt.scatter(w0[n:,n::step3],weMF[n:,n::step3],marker='^',color='r',facecolors='none',s=6,label='eMF',zorder=2)
    plt.scatter(w0[n:,n::step3],wMLE[n:,n::step3],marker='o',color='b',facecolors='none',s=6,label='MLE',zorder=3)
    plt.scatter(w0[n:,n::step3],wFEM[n:,n::step3],marker='o',color='k',s=6,label='FEM',zorder=4)
    plt.xlabel('actual interactions',labelpad=0)
    plt.ylabel('predicted interactions',labelpad=0)
    plt.xlim([wmin,wmax])
    plt.ylim([wmin,wmax])


#================================================================================================================
def plot_MSE(ifig):
    ext_name2 = '%1.1f_%02d'%(g,ln)

    nMF = np.loadtxt('MSE_nMF_%s.dat'%ext_name2)
    TAP = np.loadtxt('MSE_TAP_%s.dat'%ext_name2)
    eMF = np.loadtxt('MSE_eMF_%s.dat'%ext_name2)
    MLE = np.loadtxt('MSE_MLE_%s.dat'%ext_name2)
    FEM = np.loadtxt('MSE_FEM_%s.dat'%ext_name2)

    pv = FEM[:,1]
    print(pv)

    MSEmin = 0.0002
    MSEmax = 0.2
    pvmin = 59.7
    pvmax = 90.3

    labels = ['I','J','K','L']
    #titles = ['observed to observed','hidden to observed','observed to hidden','hidden to hidden']
    # obs to obs
    for i in range(4):    
        plt.subplot2grid((nrow,ncol),(ifig,i))
        #plt.title(titles[i])
        plt.text(54,0.2,labels[i],fontsize=13)
        plt.semilogy(pv,nMF[:,i+2],'gx--',linewidth=1,markersize=4,label='nMF')
        plt.semilogy(pv,TAP[:,i+2],'cD--',linewidth=1,markersize=4,markerfacecolor='None',label='TAP')
        plt.semilogy(pv,eMF[:,i+2],'r^--',linewidth=1,markersize=4,label='eMF')
        plt.semilogy(pv,MLE[:,i+2],'bo--',linewidth=1,markersize=4,markerfacecolor='None',label='MLE')
        plt.semilogy(pv,FEM[:,i+2],'ko-',linewidth=1,markersize=4,label='FEM')
        plt.xlabel('percent visible')
        plt.ylabel('MSE')
        plt.xticks([60,70,80,90],('60%','70%','80%','90%'))
        plt.xlim([pvmin,pvmax])
        if i ==0: plt.legend(prop={'size': 7},labelspacing=0.25,loc='lower right',bbox_to_anchor=(1.,0.13))
        plt.ylim([MSEmin,MSEmax])
#==================================================================================================================

nh0_list = [10,40]
for ifig,nh0 in enumerate(nh0_list):
    plot_w(nh0,ifig)    
    
plot_MSE(2)
    
plt.tight_layout(h_pad=1, w_pad=1.5)
#plt.savefig('fig4_all_v2.pdf', format='pdf', dpi=100)
plt.savefig('fig4_all_v2.eps', format='eps', dpi=100)
```
