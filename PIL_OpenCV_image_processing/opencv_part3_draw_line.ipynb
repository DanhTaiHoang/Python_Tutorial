```python
import cv2 as cv

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## read
a = cv.imread("image_test.png")
b = cv.cvtColor(a, cv.COLOR_BGR2GRAY)
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,6))
ax[0].imshow(a)
ax[1].imshow(b, cmap="gray")
```

```python
c = a.copy()
```

```python
c[:,:,0] = b
c[:,:,1] = b
c[:,:,2] = b
```

```python
plt.imshow(c)
```

```python
x1 = 100
y1 = 200
x2 = 300
y2 = 400
line_thickness = 2
c = cv.line(c, (x1, y1), (x2, y2), (0, 255, 0), thickness=line_thickness)
```

```python
plt.imshow(c)
```

```python
mask_tile_size = 100
c[:, ::mask_tile_size, :] = [0,0,255]
c[:: mask_tile_size, :, :] = [0,0,255]
plt.imshow(c)
```
