```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Legends title and labels

x = [1, 2, 3]
y = [3, 5, 2]
y2 = [4, -1., 3]
y3 = [1, 4., -2]
plt.title('Linear couplings')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim([0.95,3.05])
plt.ylim([-2,6])
plt.grid()
plt.xticks(np.arange(1,3.5,step=0.5))
plt.yticks(np.arange(-2,6,step=2),('low','medium','high','very high'),rotation='30')
plt.plot(x, y, label='y',marker='^',color='k')
plt.plot(x, y2, label='y2',marker='o',color='r')
plt.plot(x, y3, label='y3',marker='s',color='b',linestyle='--',linewidth=1)
plt.legend()
plt.show()
```

```python
# Bar Charts
x = [1, 3, 5, 7, 9]
y = [3, 5, 2, 4, 6]

x2 = [2, 4, 6, 8, 10]
y2 = [4, 2, 3, 1, 5]

plt.bar(x, y, label='y', color='b')
plt.bar(x2, y2, label='y2', color='r')
plt.title('Bar charts')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Histogram:
mu, sigma,N = 0, 0.1, 1000 # mean, standard deviation and number of points
x = np.random.normal(mu,sigma,N)
bins = np.linspace(min(x), max(x),10, endpoint=False)
plt.hist(x,bins,histtype='bar',rwidth=0.8)
plt.title('Histogram')
plt.xlabel('X')
plt.ylabel('P')
#plt.legend()
plt.show()
```

```python
# Scatter:

x = [1, 2, 3, 3, 2, 5, 7, 6, 4]
y = [2, 5, 1, 4, 3, 2, 8, 7, 5]

plt.title('Scatter plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x, y, label='y',marker='*',s=30) # s: marker size
plt.show()
```

```python
# Stack:
x = [1, 2, 3, 4]
y1 = [3, 2, 4, 6]
y2 = [1, 1, 3, 4]
y3 = [2, 1, 2, 5]

plt.stackplot(x, y1,y2,y3, colors=['k','b','r'])

# create legend:
plt.plot([], [], color='k', label='y1', linewidth=5)
plt.plot([], [], color='b', label='y2', linewidth=5)
plt.plot([], [], color='r', label='y3', linewidth=5)

plt.title('Stack plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Multiple axis
# Plot y1 vs x on the left axis:
x = [1, 2, 3]
y1 = [3, 5, 2]
y2 = [4, -1., 3]

plt.xlabel("x")
plt.ylabel("y1", color="b")
plt.tick_params(axis="y", labelcolor="b")
plt.plot(x, y1, "bo-", linewidth=2)

# Plot y2 vs x on the right axis:
plt.twinx()
plt.ylabel("y2", color="r")
plt.tick_params(axis="y", labelcolor="r")
plt.plot(x, y2, "r^--", linewidth=2)

plt.show()
```

```python
# Display minor ticks:
# Plot y1 vs x on the left axis:
from matplotlib.ticker import AutoMinorLocator

x = [1, 2, 3]
y1 = [3, 5, 2]
y2 = [4, -1., 3]

ax = plt.subplot(1, 1, 1)
plt.xlabel("x")
plt.ylabel("y1", color="b")
ax.yaxis.set_minor_locator(AutoMinorLocator(5)) # minor ticks
plt.tick_params(axis="y", labelcolor="b")
plt.plot(x, y1, "bo-", linewidth=2)

# Plot y2 vs x on the right axis:
ax2 = plt.twinx()
plt.ylabel("y2", color="r")
plt.tick_params(axis="y", labelcolor="r")
ax2.yaxis.set_minor_locator(AutoMinorLocator(2)) # minor ticks
plt.plot(x, y2, "r^--", linewidth=2)

plt.show()
```

```python
# multiple plots
x = [1, 2, 3]
y1 = [3, 5, 2]
y2 = [4, -1., 3]

fig, ax = plt.subplots(2, 2)
ax[0,0].scatter(x,y1)
ax[0,1].plot(x,y2)

plt.tight_layout(h_pad=0.5, w_pad=1)
plt.show()
```

```python
# multiple plots
x = [1, 2, 3]
y1 = [3, 5, 2]
y2 = [4, -1., 3]
 
plt.subplot(2, 1, 1) # number of row, column, and figure index
plt.plot(x, y1, 'ko-')
plt.title('2 subplots')
plt.ylabel('neuron 1')

plt.subplot(2, 1, 2)
plt.plot(x, y2, 'b.-')
plt.xlabel('time (s)')
plt.ylabel('neuron 2')

plt.show()
```

```python
# multiple plots
x = [1, 2, 3]
y1 = [3, 5, 2]
y2 = [4, -1., 3]
y3 = [2, -2., -1]
 
plt.subplot(2, 2, 1) # number of row, column, and figure index
plt.plot(x, y1, 'ko-')
plt.title('2 subplots')
plt.ylabel('neuron 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, 'b.-')
plt.xlabel('time (s)')
plt.ylabel('neuron 2')

plt.subplot(2, 1, 2)
plt.plot(x, y3, 'r.--')
plt.xlabel('time (s)')
plt.ylabel('neuron 3')

plt.tight_layout(h_pad=0.5, w_pad=1)
plt.show()
```
